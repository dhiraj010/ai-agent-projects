- **Scikit-learn**: Handles simpler AI tasks (e.g., predicting numbers, sorting data). Easy and fast.

# PyTorch Linear Regression Example
Predicts a line (y = 2x + 1) from data points using a neural network.

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# Create data: y = 2x + 1 (with some noise)
x = torch.linspace(0, 10, 100).reshape(-1, 1)  # 100 points from 0 to 10
y = 2 * x + 1 + torch.randn(x.size()) * 0.5  # y = 2x + 1 + small random noise

# Define a simple neural network (linear model)
model = nn.Linear(1, 1)  # 1 input (x), 1 output (y)

# Set up loss function and optimizer
loss_fn = nn.MSELoss()  # Measures how wrong the predictions are
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)  # Adjusts the model to reduce errors

# Train the model
for step in range(100):  # 100 training steps
    y_pred = model(x)  # Predict y
    loss = loss_fn(y_pred, y)  # Calculate error
    optimizer.zero_grad()  # Clear old adjustments
    loss.backward()  # Figure out how to improve
    optimizer.step()  # Update the model

# Get the learned line (y = mx + b)
m = model.weight.item()  # Slope
b = model.bias.item()  # Intercept
print(f"Learned line: y = {m:.2f}x + {b:.2f}")

# Plot the data and learned line
plt.scatter(x, y, label="Data points")
plt.plot(x, model(x).detach(), 'r-', label="Learned line")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

## How the Example Works
- **Goal**: Predict a line (y = 2x + 1) from scattered data points.
- **Data**: 100 points where y is roughly 2x + 1 (with some randomness).
- **Model**: A simple neural network (one line) that learns the slope (m) and intercept (b).
- **Training**: The model guesses, checks its errors, and improves 100 times.
- **Result**: A line close to y = 2x + 1 (e.g., y = 2.03x + 0.95) and a plot showing the fit.
- **Insight**: PyTorch learns patterns (like lines) by practicing, which is how AI works for bigger tasks.
