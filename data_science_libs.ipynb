```python
import numpy as np
arr = np.array([1, 2, 3, 4, 5])
print(arr)
print(arr.shape)    # Shape (how many elements)
print(arr.dtype)    # Data type of elements
print(arr.ndim)     # Number of dimensions (1D, 2D, etc.)
print(arr + 10)     # Adds 10 to each element
print(arr * 2)      # Multiplies each element by 2
print(arr ** 2)     # Squares each element
np.zeros(5)         # [0. 0. 0. 0. 0.]
np.ones(5)          # [1. 1. 1. 1. 1.]
np.arange(0, 10, 2) # [0 2 4 6 8]
np.linspace(0, 1, 5) # [0.   0.25 0.5  0.75 1.  ]
print(arr[0])     # First element
print(arr[1:4])   # Slice from index 1 to 3
arr[0] = 100      # Modify element
print(arr)



```

# Numpy : Working with numbers

```python
import numpy as np
# Make a list of numbers (array)
numbers = np.array([1, 2, 3, 4])
# Do some math
total = numbers.sum()  # Add them up
average = numbers.mean()  # Find the average
doubled = numbers * 2  # Double each number
# Pick one number
first = numbers[0]
# Show results
print("Numbers:", numbers)
print("Total:", total)
print("Average:", average)
print("Doubled:", doubled)
print("First number:", first)
```

# Pandas : Working with tables

```python
import pandas as pd
# Make a table (DataFrame) with names and ages
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35]
}
table = pd.DataFrame(data)
# Find the average age
avg_age = table["Age"].mean()
# Pick people over 28
older = table[table["Age"] > 28]
# Show results
print("Table:\n", table)
print("Average Age:", avg_age)
print("People over 28:\n", older)
```

# matplotlib : making charts

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Make data for a line chart
x = np.linspace(0, 10, 100)  # 100 numbers from 0 to 10
y = np.sin(x)  # Sine wave

# Draw a line chart
plt.plot(x, y, 'b-', label="Sine Wave")
plt.title("Sine Wave")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```
